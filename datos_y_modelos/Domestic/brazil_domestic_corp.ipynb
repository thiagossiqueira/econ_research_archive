### 1) Leer el archivo Excel

```python
import pandas as pd
import numpy as np
from scipy.stats import kurtosis

file_path = 'brazil_domestic_corp_db.xlsx'

df = pd.read_excel(file_path, sheet_name='db_values_only')

df = df.replace(r'^#N/A.*', np.nan, regex=True)

df = df.dropna(how='all')

df
```

```python
df.to_excel('1.xlsx', index=False)
```

```python
df.columns
```

### 2) en industry quitar los que son estatales (estados varios)

```python
df = df[~df['CLASSIFICATION_LEVEL_4_NAME'].str.startswith("Government", na=False)]
df
```

```python
df.to_excel('2.xlsx', index=False)
```

### 2.1) en industry quitar los que son financeiros (eg. banks)

```python
df = df[~df['industry_sector'].isin(['Financial'])]
df
```

### 3.1) Total de emisores

```python
print(f"Total de emisores: {df['ISSUER'].nunique()}")
```

### FunciÃ³n de EstadÃ­sticas descriptivas sobre la muestra filtrada

```python
def descriptive_stats(col):
    return {
        'mean': col.mean(),
        'min': col.min(),
        'max': col.max(),
        'median': col.median(),
        'std_dev': col.std(),
        'kurtosis': kurtosis(col.dropna())
    }
```

### 4) luego agrupar por empresa los bonos que emitiÃ³ cada uno y ahi vemos cuantos hay de cada una , con que maturity, volume issued y sobre todo volume traded asÃ­ elegimos los mÃ¡s lÃ­quidos

```python
grouped = df.groupby(by=['ISSUER'])

issuances_count = grouped[["ID_ISIN"]].count().rename(columns={"ID_ISIN": "NUMBER_OF_ISSUANCES"})

maturity_min = grouped[["MATURITY"]].min().rename(columns={"MATURITY": "MATURITY_MIN"})
maturity_max = grouped[["MATURITY"]].max().rename(columns={"MATURITY": "MATURITY_MAX"})

volume_issued_sum = grouped[["AMT_ISSUED"]].sum().rename(columns={"AMT_ISSUED": "VOLUME_ISSUED_SUM"})
volume_issued_min = grouped[["AMT_ISSUED"]].min().rename(columns={"AMT_ISSUED": "VOLUME_ISSUED_MIN"})
volume_issued_max = grouped[["AMT_ISSUED"]].max().rename(columns={"AMT_ISSUED": "VOLUME_ISSUED_MAX"})
volume_issued_avg = grouped[["AMT_ISSUED"]].median().rename(columns={"AMT_ISSUED": "VOLUME_ISSUED_AVG"})


number_of_trades_min = grouped[["LQA_EMPIRICAL_TRADE_COUNT"]].min().rename(columns={"LQA_EMPIRICAL_TRADE_COUNT": "NUMBER_OF_TRADEDS_MIN"})
number_of_trades_max = grouped[["LQA_EMPIRICAL_TRADE_COUNT"]].max().rename(columns={"LQA_EMPIRICAL_TRADE_COUNT": "NUMBER_OF_TRADEDS_MAX"})
number_of_trades_avg = grouped[["LQA_EMPIRICAL_TRADE_COUNT"]].median().rename(columns={"LQA_EMPIRICAL_TRADE_COUNT": "NUMBER_OF_TRADEDS_AVG"})

   

maturity_summary = pd.concat([issuances_count, 
                              maturity_min, 
                              maturity_max, 
                              volume_issued_sum, 
                              volume_issued_min, 
                              volume_issued_max, 
                              volume_issued_avg,
                              number_of_trades_min,
                              number_of_trades_max,
                              number_of_trades_avg,
                             ], axis=1).reset_index()
   


maturity_summary = maturity_summary.sort_values(by="NUMBER_OF_ISSUANCES", ascending=False)

maturity_summary
```

```python
maturity_summary.to_excel('4.xlsx', index=False)
```

### 4.1.2) liquidity = market turn over / market value outstanding 

```python
df['calculated_liquidity'] = df['LQA_EMPIRICAL_TOTAL_TRADE_VOLUME'] / df['AMT_OUTSTANDING']
df
```

### 4.2) elegimos los mÃ¡s lÃ­quidos (trade volume mayor que la media)

```python
# Filter top-liquid bonds (above median)
top_liquid = df[df['calculated_liquidity'] > df['calculated_liquidity'].median()]

# Sort descending by liquidity
top_liquid = top_liquid.sort_values(by='calculated_liquidity', ascending=False)

top_liquid
```

```python
top_liquid.to_excel('4.2.xlsx', index=False)
```

### 4.2.1) filtrada esta muestra haces las estadÃ­sticas descriptivas incluyendo media, max min, median, std dev, kurtosis para los numeros de negocios

```python
stats_trade_count = descriptive_stats(top_liquid['LQA_EMPIRICAL_TRADE_COUNT'])
stats_trade_count
```

### 4.2.2) filtrada esta muestra haces las estadÃ­sticas descriptivas incluyendo media, max min, median, std dev, kurtosis para los volume de negocios

```python
stats_trade_volume = descriptive_stats(top_liquid['LQA_EMPIRICAL_TOTAL_TRADE_VOLUME'])
stats_trade_volume
```
